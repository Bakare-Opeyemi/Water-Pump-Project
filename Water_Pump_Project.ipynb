```python
#Creating a function for plotting columns in the dataset
def plotColumn(columns,titles):
    axes= []
    for i in range(len(columns)):
        axes.append(columns[i].plot(figsize=(10,5)))
    for j in range(len(axes)):
        axes[j].set_title(titles[j])
    plt.show()
```

```python
columns = [data[data.columns[:5]],data[data.columns[5:10]],data[data.columns[10:15]]]
titles = ['sensor00-sensor04','sensor05-sensor09','sensor10-sensor14']
plotColumn(columns,titles)
```

<p> To get deeper insights into the data, we will plot the percent changes of some sensor values in relation to the target variable </p>

```python
changes= data.pct_change().dropna()
columns = [changes[changes.columns[:5]],changes[changes.columns[5:10]],changes[changes.columns[10:15]]]
titles = ['Changes in sensor00-sensor04 readings','Changes in sensor05-sensor09 readings','Changes in  sensor10-sensor14 readings']
plotColumn(columns,titles)
```

```python
#Creating a function for plotting columns in the dataset
def plotColumn(columns,titles):
    axes= []
    for i in range(len(columns)):
        axes.append(columns[i].plot(figsize=(10,5)))
    for j in range(len(axes)):
        axes[j].set_title(titles[j])
        axes[j].axhspan(-1,1, color ='violet',alpha = 0.3)
    plt.show()
```

```python
columns = [changes[['sensor_10', 'sensor_16', 'rul']],changes[['sensor_20','sensor_27','rul']],changes[['sensor_30','sensor_35','rul']]]
titles = ['Changes in sensor_10,15 and Remaining Useful Life','Changes in sensor_20,27 and Remaining Useful Life','Changes in  sensor30,35 and Remaining Useful Life']
plotColumn(columns,titles)
```

### INFERENCE

<p>from the plots, we can see that the changes in sensor readings over a period of time is relatively small or near zero with very high spikes at random intervals</p>

```python
#Creating a function for plotting aggregates of features in the dataset
def plotAggregates(columns,titles):
    axes= []
    for i in range(len(columns)):
        axes.append(columns[i].plot(figsize=(12,5)))
    for j in range(len(axes)):
        axes[j].set_title(titles[j])
    plt.show()
```

```python
columns = [data_by_month[data_by_month.columns[9:12]],data_by_month[data_by_month.columns[12:15]],data_by_month[data_by_month.columns[15:18]]]
titles = ['Monthly Aggregate of  sensor09, 10 and 11 Data ','Monthly Aggregate of  sensor 12, 13 and 14 Data ','Monthly Aggregate of  sensor16, 17 and 18 Data: CASE-STUDY B1 ']
plotAggregates(columns,titles)
```

```python
columns = [data_by_month[data_by_month.columns[18:21]],data_by_month[data_by_month.columns[21:25]],data_by_month[data_by_month.columns[25:28]]]
titles = ['Monthly Aggregate of  sensor19, 20 and 21 Data:CASE-STUDY B2 ','Monthly Aggregate of  sensor 22,23,24 and 25 Data:CASE-STUDY C ','Monthly Aggregate of  sensor 26,27 and 28 Data ']
plotAggregates(columns,titles)
```

```python
columns = [data_by_month[data_by_month.columns[28:32]],data_by_month[data_by_month.columns[32:36]],data_by_month[data_by_month.columns[36:40]]]
titles = ['Monthly Aggregate of  sensor29, 30, 31 and 32 Data: CASE-STUDY C1','Monthly Aggregate of  sensor 33, 34, 35 and 36 Data ','Monthly Aggregate of  sensor 37,38,39 and 40 Data: CASE-STUDY D ']
plotAggregates(columns,titles)
```

```python
columns = [data_by_month[data_by_month.columns[40:44]],data_by_month[data_by_month.columns[44:48]],data_by_month[data_by_month.columns[48:50]]]
titles = ['Monthly Aggregate of  sensor41, 42, 43 and 44 Data: CASE-STUDY E ','Monthly Aggregate of  sensor 45, 46, 47 and 48 Data ','Monthly Aggregate of  sensor 49 and 51 Data ']
plotAggregates(columns,titles)
```

```python
plotAggregates([data['rul']],['Monthly Aggregate of the Remaining Useful life of The water Pump'])
```

### INFERENCE
<p>From the monthly aggregate plots of features in the dataset, the following patterns have been observed; </p>
<ul>
    <li>Most sensors record lower values in June or sometime between May and June</li>
    <li>The plots also highlight the relationship between highly correlating sensors such as seen in CASE-STUDIES A,C,D and E. Here we can see that some plots are simply approximates of each other and it is very possible to predict one plot from the other based on this monthly trend.</li>
    <li>We can also see that many sensors record their highest values in May and July. This is evident in CASE-STUDIES C1,B1,B2 and B </li>
</ul>
